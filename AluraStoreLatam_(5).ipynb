### 1. Análisis de facturación y costos de envío promedio por tienda

```python
# Paso 1: Importar pandas
import pandas as pd

# Paso 2: Cargar los datos desde GitHub
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Paso 3: Calcular métricas
precio_total_tiendas = [sum(tienda1['Precio']), sum(tienda2['Precio']), sum(tienda3['Precio']), sum(tienda4['Precio'])]
precio_total_promedio = [round(i / len(tienda1['Calificación']), 2) for i in precio_total_tiendas]

costo_envio = [sum(tienda1['Costo de envío']), sum(tienda2['Costo de envío']), sum(tienda3['Costo de envío']), sum(tienda4['Costo de envío'])]
costo_envio_promedio = [round(i / len(tienda1['Calificación']), 2) for i in costo_envio]

total_costos = [precio_total_tiendas[i] - costo_envio[i] for i in range(4)]

# Paso 4: Porcentaje de ventas y ganancias
total_ventas = sum(precio_total_tiendas)
total_ganancias = sum(total_costos)

porcentaje_ventas = [round((v / total_ventas) * 100, 2) for v in precio_total_tiendas]
porcentaje_ganancias = [round((g / total_ganancias) * 100, 2) for g in total_costos]

# Paso 5: Mostrar resultados
for i in range(4):
    print(f"Tienda {i+1}:")
    print(f"  Ventas totales: ${precio_total_tiendas[i]:,.2f}")
    print(f"  Promedio precio: ${precio_total_promedio[i]:,.2f}")
    print(f"  Costo envío total: ${costo_envio[i]:,.2f}")
    print(f"  Promedio envío: ${costo_envio_promedio[i]:,.2f}")
    print(f"  Ganancia neta: ${total_costos[i]:,.2f}")
    print(f"  % del total de ventas: {porcentaje_ventas[i]}%")
    print(f"  % del total de ganancias: {porcentaje_ganancias[i]}%")
    print("-" * 40)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Nombres de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# DataFrame con resultados
df_resultados = pd.DataFrame({
    'Tienda': tiendas,
    'Ventas Totales': precio_total_tiendas,
    'Ganancia Neta': total_costos
})

# Gráfico de barras
ax = df_resultados.plot(kind='bar', x='Tienda', color=['#4CAF50', '#2196F3'], edgecolor='black')
plt.title('Comparación de Ventas Totales y Ganancia Neta por Tienda', fontsize=14)
plt.ylabel('Monto en Pesos ($)', fontsize=12)
plt.xlabel('Tiendas', fontsize=12)
plt.xticks(rotation=0)
plt.legend(loc='upper right')

# Etiquetas centradas en las barras
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='center', fontsize=10, color='white')

plt.tight_layout()
plt.show()

# Gráfico circular del porcentaje de ventas
plt.figure(figsize=(8, 6))
plt.pie(porcentaje_ventas, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Distribución del Porcentaje de Ventas Totales por Tienda', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Limpiar nombres de columnas
for tienda in [tienda1, tienda2, tienda3, tienda4]:
    tienda.columns = tienda.columns.str.strip()

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todas las categorías
todas_categorias = sorted(set().union(*[tienda['Categoría del Producto'].unique() for tienda in tiendas]))

# Crear DataFrames para ventas en pesos y porcentaje por categoría
ventas_cop = pd.DataFrame(index=todas_categorias, columns=nombres_tiendas)
ventas_pct = pd.DataFrame(index=todas_categorias, columns=nombres_tiendas)

for i, tienda in enumerate(tiendas):
    ventas_por_categoria = tienda.groupby('Categoría del Producto')['Precio'].sum()
    total_ventas = ventas_por_categoria.sum()
    for cat in todas_categorias:
        valor = ventas_por_categoria.get(cat, 0)
        ventas_cop.iloc[todas_categorias.index(cat), i] = valor
        ventas_pct.iloc[todas_categorias.index(cat), i] = round((valor / total_ventas) * 100, 1)

# Convertir a float
ventas_cop = ventas_cop.astype(float)
ventas_pct = ventas_pct.astype(float)

# Crear gráfico combinado
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, tienda in enumerate(nombres_tiendas):
    ax = axes[i]

    bars = ax.bar(ventas_cop.index, ventas_cop[tienda], color='skyblue', label='COP Vendido')
    ax.set_ylabel('Ventas en COP')
    ax.set_title(f'Ventas por Categoría - {tienda}', fontsize=12)
    ax.tick_params(axis='x', rotation=45)

    # Agregar porcentaje con eje secundario
    ax2 = ax.twinx()
    ax2.plot(ventas_pct.index, ventas_pct[tienda], color='orange', marker='o', label='% de Ventas')
    ax2.set_ylabel('% del Total de Ventas')

    # Etiquetas de valores
    for bar, p in zip(bars, ventas_pct[tienda]):
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2, height + 5000, f"${int(height):,}", ha='center', fontsize=8)
            ax2.text(bar.get_x() + bar.get_width()/2, p + 1, f"{p}%", ha='center', va='bottom', fontsize=8, color='darkorange')

# Layout final
fig.suptitle('Ventas por Categoría en Pesos y Porcentaje (%)', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# URLs de los archivos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Limpiar nombres de columnas
for tienda in [tienda1, tienda2, tienda3, tienda4]:
    tienda.columns = tienda.columns.str.strip()

# Lista de tiendas
tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todas las categorías
categorias = sorted(set().union(*[t['Categoría del Producto'].unique() for t in tiendas]))

# Crear DataFrame consolidado
ventas_categoria = pd.DataFrame(index=categorias, columns=nombres_tiendas)

# Llenar el DataFrame
for i, tienda in enumerate(tiendas):
    ventas = tienda.groupby('Categoría del Producto')['Precio'].sum()
    for categoria in categorias:
        ventas_categoria.iloc[categorias.index(categoria), i] = ventas.get(categoria, 0)

# Asegurar tipo numérico
ventas_categoria = ventas_categoria.fillna(0).astype(float)

# Gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(14, 7))

# Parámetros de posición
x = np.arange(len(categorias))
width = 0.2

# Dibujar cada barra
for i, tienda in enumerate(nombres_tiendas):
    ax.bar(x + i*width, ventas_categoria[tienda], width, label=tienda)

# Etiquetas
ax.set_xlabel('Categorías de Productos')
ax.set_ylabel('Ventas en COP')
ax.set_title('Comparativo de Ventas por Categoría entre Tiendas')
ax.set_xticks(x + width*1.5)
ax.set_xticklabels(categorias, rotation=45, ha='right')
ax.legend()

# Agregar etiquetas encima de las barras
for i, tienda in enumerate(nombres_tiendas):
    for j, valor in enumerate(ventas_categoria[tienda]):
        ax.text(x[j] + i*width, valor + 2000, f"${int(valor):,}", ha='center', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Limpiar nombres de columnas
for tienda in [tienda1, tienda2, tienda3, tienda4]:
    tienda.columns = tienda.columns.str.strip()

# Función para contar métodos de pago por tienda
def contar_metodos(tienda, nombre_tienda):
    conteo = tienda['Método de pago'].value_counts()
    return conteo.rename(nombre_tienda)

# Contar métodos por tienda
metodos1 = contar_metodos(tienda1, 'Tienda 1')
metodos2 = contar_metodos(tienda2, 'Tienda 2')
metodos3 = contar_metodos(tienda3, 'Tienda 3')
metodos4 = contar_metodos(tienda4, 'Tienda 4')

# Unir todos en un solo DataFrame
df_metodos = pd.concat([metodos1, metodos2, metodos3, metodos4], axis=1).fillna(0).astype(int)

# Gráfico de barras horizontales
df_metodos.plot(kind='barh', figsize=(10, 6), colormap='Pastel1')

# Personalizar
plt.title('Métodos de Pago por Tienda')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Método de Pago')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos desde URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Limpiar nombres de columnas
for tienda in [tienda1, tienda2, tienda3, tienda4]:
    tienda.columns = tienda.columns.str.strip()

# Lista de tiendas
tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todos los métodos de pago únicos
todos_metodos = set()
for tienda in tiendas:
    todos_metodos.update(tienda['Método de pago'].unique())
todos_metodos = sorted(todos_metodos)

# Preparar datos por tienda
porcentajes_por_tienda = []
for tienda in tiendas:
    total = len(tienda)
    porcentaje = tienda['Método de pago'].value_counts(normalize=True) * 100
    porcentajes = [porcentaje.get(metodo, 0) for metodo in todos_metodos]
    porcentajes_por_tienda.append(porcentajes)

# Transponer para graficar por método
import numpy as np

porcentajes_array = np.array(porcentajes_por_tienda).T
bar_width = 0.6
colors = plt.cm.Set3.colors  # Colores distintos

# Crear gráfico
fig, ax = plt.subplots(figsize=(10, 6))

bottom = np.zeros(len(nombres_tiendas))

for i, metodo in enumerate(todos_metodos):
    valores = porcentajes_array[i]
    bars = ax.bar(nombres_tiendas, valores, bottom=bottom, label=metodo, color=colors[i % len(colors)])

    # Etiquetas de porcentaje en cada segmento
    for j, bar in enumerate(bars):
        height = bar.get_height()
        if height > 3:  # solo mostrar si es significativo
            ax.text(bar.get_x() + bar.get_width()/2, bottom[j] + height/2, f'{height:.1f}%',
                    ha='center', va='center', fontsize=9, color='black')

    bottom += valores

# Configuraciones
ax.set_ylabel('Porcentaje de uso (%)')
ax.set_title('Distribución de Métodos de Pago por Tienda')
ax.legend(title="Método de Pago", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
